```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats
import pandas as pd
import statsmodels.api as sm

import warnings
warnings.simplefilter('ignore')

import get_project_data
```

```python
df = get_project_data.df
```

```python
df
```

```python
# from LEO
# Group by month:
n_days = len(df)

months_from_june = []
for k in range(n_days):
    months_from_june.append(6 - np.abs(int(df["Datum und Uhrzeit"][k][3:5])-6))


# Plotting:
fig, ax = plt.subplots()

sc = plt.scatter(df[" SDK"], df["Gesamtanlage[kWh]"], c=months_from_june, cmap='jet')

cbar = plt.colorbar(sc)
cbar.ax.get_yaxis().set_ticks([])

for j, month in enumerate(['Dec','Nov/Jan','Oct/Feb','Sep/Mar','Aug/Apr','Jul/May','Jun']):
    cbar.ax.text(1.5, j, month, ha='left', va='center')
    #cbar.ax.text(1.5, 6/7*(j + .5), month, ha='left', va='center')

plt.xlabel("Hours of sunshine")
plt.ylabel("Electricity produced in kWh")

plt.show()
```

```python
sns.pairplot(df)
plt.savefig("pairwise_corr.png")
plt.show()
```

```python
# pairwise correlation
pairwise_corr = {}
for feature in df.columns[1:]:
    pairwise_corr[feature] = []
    for second_feature in df.columns[1:]:
        res = stats.pearsonr(df[feature], df[second_feature])
        pairwise_corr[feature].append(res.statistic)
pairwise_corr_df = pd.DataFrame.from_dict(pairwise_corr, orient='index', columns=df.columns[1:])
pairwise_corr_df
```

**production in kwH**: highest correlation with sun shine hours (0.81), temperature features, snow and form of rainfall

```python
model = sm.GLM(endog=df["Gesamtanlage[kWh]"], exog=df[df.columns[2:]])
results = model.fit()
print(results.summary())
```

```python
model = sm.OLS(endog=df["Gesamtanlage[kWh]"], exog=df[df.columns[2:]])
results = model.fit()
print(results.summary())
```
